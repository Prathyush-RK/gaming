```python
import cv2
```

```python
cap=cv2.VideoCapture("video.mp4")
fgbg=cv2.createBackgroundSubtractorMOG2()
```

```python
while True:
    ret, frame=cap.read()
    if not ret:
        break
    mask=fgbg.apply(frame)
    cv2.imshow("Video",frame)
    cv2.imshow("Motion Mask",mask)
    if cv2.waitKey(30) & 0xFF==ord('q'):
        break
```
