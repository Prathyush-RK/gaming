```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Read and Display Image

img=cv2.imread("input.jpg",cv2.IMREAD_GRAYSCALE)
print(img)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Display Image Attributes
img=cv2.imread("input.jpg",cv2.IMREAD_GRAYSCALE)
print("Shape of image (h,w): ",img.shape)
print("Datatype of image: ",img.dtype)
print("Size of image (in pixels): ",img.size)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Display Colour Attributes
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
print("Shape of image (h,w,channels): ",img.shape)
print("Size of image (in pixels): ",img.size)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Read and Display Colour Image
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
#Reverse channel of image (BGR to RGB Manual Method)
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
reverse=img[:,:,::-1]
plt.imshow(reverse)
plt.axis("off")
plt.show()
```

```python
#Splitting color channel of image
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
b, g, r= cv2.split(img)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(r, cmap='gray')
plt.title("Red Channel")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title("Green Channel")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(b, cmap='gray')
plt.title("Blue Channel")
plt.axis("off")

plt.show()
```

```python
#BGR to RGB Conversion
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
plt.show()
```

```python
#BGR to HSV Conversion
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.axis("off")
plt.show()
```

```python
#BGR to Gray Conversion
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.axis("off")
plt.show()
```
