```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Cropping image
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
crop=img[50:400, 100:600]
plt.imshow(crop)
plt.axis("off")
plt.show()
```

```python
#Resize image
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
res=cv2.resize(img,(600,400))
plt.imshow(res)
plt.axis("off")
plt.show()
```

```python
#Translation
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows, cols = img.shape[:2]
M=np.float32([[1,0,30],
             [0,1,20]])
#tx - 30 ty - 20
translate=cv2.warpAffine(img,M,(cols,rows))
plt.imshow(translate)
plt.axis("off")
plt.show()
```

```python
#Rotation
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows, cols = img.shape[:2]
center=(rows//2,cols//2)
angle=45
scale=1
M=cv2.getRotationMatrix2D(center,angle,scale)
rotate=cv2.warpAffine(img,M, (cols,rows))
plt.imshow(rotate)
plt.axis("off")
plt.show()
```

```python
#Shearing
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows, cols = img.shape[:2]
M=np.float32([[1,0.5,0],
             [0,1,0]])
#0.5 - Shearing factor
shear=cv2.warpAffine(img,M,(int(1.5*cols),rows))
plt.imshow(shear)
plt.axis("off")
plt.show()
```

```python
#Scaling (Resize image)
img=cv2.imread("input.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
res=cv2.resize(img,(1000,800))
plt.imshow(res)
plt.axis("off")
plt.show()
```
